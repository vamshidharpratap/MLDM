```python
#a. To delete duplicate samples. - 

def del_duplicate(data: pandas.core.frame.DataFrame) -> pandas.core.frame.DataFrame:
    '''
    The function del_duplicate will delete the duplicate. 
    param data: DataFrame as an input 
    
    return : DataFrame with no duplicate values
    '''
    
    # Using pandas drop_duplicates() dataframe function that removes the duplicates and retures the dataframe 
    data_with_no_duplicates = data.drop_duplicates()
    
    return data_with_no_duplicates
```

```python
#b. Function to fill missing values (fields that have 0) with the mean of the corresponding class of the attribute

def fill_values(data: pandas.core.frame.DataFrame) -> pandas.core.frame.DataFrame :
    '''
    This function fill_values will take the input pandas dataframe and the corresponding missing values in 
    the data frame are filled with the mean of that attribute.
    
    param data: DataFrame as an input 
    return : DataFrame with missing values imputed by mean of that attribute

    '''
    # we will be using pandas replace function
    data = data.replace(to_replace = 0 , value = data.mean(axis = 0)) # axis=0 means column (attribute) 
    return data
    
```

```python
#c. Function to transform nominal attributes (not the classes!) into numerical values. 
   
def transform_nominal(data: pandas.core.frame.DataFrame) -> pandas.core.frame.DataFrame :
    '''
    This function transforms the nominal attributes to numerical values by adding a new column for each nominal value
    which contains numerical values
    
    param data: DataFrame as an input 
    return : DataFrame as an output 
    '''
    #we use get_dummies() function which is provided by pandas library. This perfoms One hot encoding on the column
    new_coumns = pd.get_dummies(data.Colour)
    
    #merging the dataframe with the new coloumns.we can use join() function
    new_data = data.join(new_coumns)
    
    #Deleting the column name 'Colour '
    new_data = new_data.drop(labels = 'Colour', axis = 1)
    
    return new_data
```

```python
#d. Function to normalise the values of each numerical attribute with min-max method

def normalise_minmax(data: pandas.core.frame.DataFrame) -> pandas.core.frame.DataFrame :
    '''
    This function performs min max normalization on the dataframe that has been given as the input and returns the 
    dataframe as an output that has normalized values
    
    param data: DataFrame as an input 
    return : DataFrame as an output 
    
    '''  
    #select the columns which are numeric 
    
    numerics_toselect = ['float32', 'float64', 'int32', 'int64']
    column_names = data.select_dtypes(include=numerics_toselect).columns.to_numpy()
    
    for name in column_names:
        data[name] = (data[name]-data[name].min())/(data[name].max()-data[name].min())
    return data    
```
