```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Downloads//Country-data.csv")
```

```python
country_col = data['country']
X = data.iloc[:,1:]
X
```

```python
from sklearn.cluster import DBSCAN
DB = DBSCAN(eps=2, min_samples=5)
```

```python
# Performing the DBSCAN algorithmm on the original data. ( DATA IS NOT NORMALIZED)
DB_model = DB.fit(X)
labels = DB_model.labels_
labels
```

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)
print('Number of Clusters : ', n_clusters)
print('Number of Outliers : ', n_noise)
```

```python
# From above we observe that all the data points are considered as noise 
```
