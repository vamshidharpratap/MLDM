```python
import numpy as np
import datetime # this module is used for converting the string format of date to date format

 
```


A constructor that takes as parameters (firstname, dateOfBirth).
A method "setFirstname" that changes the firstname.
A method "getFirstname" that returns the firstname.
A method "getAge" that returns the age of the user.


```python

class User:
    
    #Defing the constructor of the class User
    def __init__(self,pFirstName, pDateOfBirth):
        self.firstname = pFirstName
        self.dateOfBirth = pDateOfBirth
        
    def setFirstname(self,pfirstname:str):
        '''
        The function setFirstname takes the parameter pfirstname and updates the firstname
        '''
        self.firstname = pfirstname
        return 
    
    def getFirstname(self):
        '''
        The function getFirstname returns the firstname.
        '''
        return self.firstname
    
    def getAge(self):
        '''
        The function getAge returns the Age of the User.
        '''
        #Converting the birthdate from string to data object. 
        born_date = datetime.datetime.strptime(self.dateOfBirth,'%m/%d/%Y').date()
        # Getting the current date by using the datetime module
        curr_date = datetime.date.today()
        diff_mon_days = (born_date.month,born_date.day) > (curr_date.month,curr_date.day)
        age = curr_date.year -  born_date.year - diff_mon_days
        return age

```

```python
# Creating the class object for the class User 
class_object = User('Vamshi','12/08/1999')
# Calling the getFirstname()
print("The firstname is :",class_object.getFirstname())

#Calling the getAge()
print("The Age of user is :",class_object.getAge())

# Calling the setFirstname()
print("\n")
print("The firstname before calling the setFirstname() function :",class_object.getFirstname())
class_object.setFirstname("Vamshidhar")
print("The firstname after calling the setFirstname() function :",class_object.getFirstname())
print("\n")
```
